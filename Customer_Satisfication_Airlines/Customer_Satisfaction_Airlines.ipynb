#### Transforming categorical variables for further analysis

```python
from sklearn.preprocessing import LabelEncoder     #Convert the categorical columns to numerical values that can be used as input for a model 
LE = LabelEncoder()
df["satisfaction_new"] = LE.fit_transform(df["satisfaction"])
df["Gender_new"] = LE.fit_transform(df["Gender"])
df["Customer Type_new"] = LE.fit_transform(df["Customer Type"]) 
df["Type of Travel_new"] = LE.fit_transform(df["Type of Travel"]) 
df["Class_new"] = LE.fit_transform(df["Class"])
df.drop(["satisfaction","Gender", "Customer Type", "Type of Travel", "Class"], axis=1, inplace=True)
df.rename(columns = {"satisfaction_new":"satisfaction","Gender_new":"Gender","Customer Type_new":"Customer Type","Type of Travel_new":"Type of Travel","Class_new":"Class"}, inplace = True)
```

#### Imputing missing data with the Simple Algorithm

```python
from sklearn.impute import SimpleImputer     
imputer = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
imputer.fit(df)
df[:] = imputer.transform(df)
```

#### Flattening continuous data

```python
from sklearn.preprocessing import KBinsDiscretizer     

Xbin = KBinsDiscretizer(n_bins=4, encode="ordinal", strategy="quantile")     #Converting 'age' attribute to categorical data of 4 values
df["Age"] = Xbin.fit_transform(df[["Age"]])
```

```python
Xbin = KBinsDiscretizer(n_bins=4, encode="ordinal", strategy="uniform")     #Converting the 'flight distance' attribute to categorical data consisting of 4 values
df["Flight Distance"] = Xbin.fit_transform(df[["Flight Distance"]])
```

```python
Xbin = KBinsDiscretizer(n_bins=4, encode="ordinal", strategy="uniform")     #Converting the 'departure delay in minutes' attribute to categorical data consisting of 4 values
df["Departure Delay in Minutes"] = Xbin.fit_transform(df[["Departure Delay in Minutes"]])
```

```python
Xbin = KBinsDiscretizer(n_bins=4, encode="ordinal", strategy="uniform")     #Converting 'arrival delay in minutes' attribute to categorical data consisting of 4 values
df["Arrival Delay in Minutes"] = Xbin.fit_transform(df[["Arrival Delay in Minutes"]])
```

#### Cleaned data ready for applying Machine learning algorithms

```python
df
```

#### Spliting the dataframe into X & y

```python
# Separate object for Predictors 
X = df.drop("satisfaction",axis=1)

# Separate object for target feature 
y = df["satisfaction"]
```

#### Train and Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 27) # Dataset is split 70% into train and 30% into test.
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

```python
#predicting test data
y_test_dt = dt_model.predict(X_test)
```

```python
# accurate prediction 
from sklearn.metrics import accuracy_score
acc_dt = accuracy_score(y_test, y_test_dt)
acc_dt
```

#### Using KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
knn_model.fit(X_train, y_train)
```

```python
#predict test data
y_test_knn = knn_model.predict(X_test)
```

```python
#accurate prediction
from sklearn.metrics import accuracy_score
acc_knn = accuracy_score(y_test, y_test_knn)
acc_knn
```
